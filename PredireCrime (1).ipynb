```python
df1 = df.drop(df[df["Y"]==90].index)
```

```python
df1=df[["DayOfWeek", "PdDistrict", "Mois", "Heure", "NumJour", "X", "Y"]]
```

```python
from category_encoders import OrdinalEncoder
```

```python
features_to_encode = [col for col in df1.columns if df1[col].dtype =='object']
encoder = OrdinalEncoder(cols=features_to_encode).fit(df1)
new_df = encoder.transform(df1)
new_df
```

```python
from sklearn.model_selection import train_test_split
```

MODELE 1

```python
X = new_df.loc[:,new_df.columns !="DayOfWeek"]
y = new_df["DayOfWeek"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=0, max_features=2)
rf.fit(X_train, y_train)
p_tr = rf.predict(X_train)
print ("Score Train -- ", round(rf.score(X_train, y_train) *100,2), " %")
```

```python
print ("Score Test -- ", round(rf.score(X_test, y_test) *100,2), " %")
```

```python
print ("Score Train -- ", round(rf.score(X_train, y_train) *100,2), " %")
```

```python
rf.feature_importances_
```

```python
pd.DataFrame(rf.feature_importances_, X_train.columns)
```
